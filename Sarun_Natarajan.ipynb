Reading Lending Club Loan Dataset

```python
import pandas as pd

df = pd.read_csv("loan.csv", low_memory=False)
df.head()
```

Checking the number of columns and rows in the dataset

```python
df.shape
```

Summary of basic statistics

```python
df.describe()
```

Missing values information

```python
df.isnull().sum()
```

```python
df.columns[df.isnull().any()].tolist()
```

Drop duplicate rows

```python
df = df.drop_duplicates()
df.shape
```

There were no duplicates rows in the dataset.

Now, checking the target variable in the dataset

```python
df['loan_status'].value_counts()
```

We will remove the "Current" rows as these are the loans which are still active and hence shouldn't be considered as they can be defaulted in future

```python
df = df[df["loan_status"] != "Current"]
df.shape
```

Removing all columns which is completely null

```python
df.dropna(axis=1, how="all", inplace=True)
df.shape
```

We are now left with 56 columns

The columns which has just 1 unique value also will have no significance to generate any insight, so removing them

```python
unique_value_count = df.nunique()
# fetching columns names which has just 1 unique value
drop_column_list = unique_value_count[unique_value_count == 1].index
df.drop(drop_column_list, axis=1, inplace=True)
df.shape
```

Features reduced to 45 columns now

Finally remove columns with very high number of null values as these cannot be imputed accurately

```python
null_columns = df.columns[df.isnull().sum() > 0]
df[null_columns].isnull().sum()
```

Columns mths_since_last_delinq and mths_since_last_record has very high null values, hence dropping them. Also desc column is a descriptive field and url column doesnt serve any purpose for analysis, hence dropping these too.

```python
high_null_count_columns = ["mths_since_last_delinq", "mths_since_last_record", "desc", "url"]
df.drop(high_null_count_columns, axis=1, inplace=True)
df.shape
```

Lets clean the int_rate column to make it useful for analysis

```python
# The % string is being removed from the value
df["int_rate"] = df["int_rate"].str.replace("%", "")
df["int_rate"] = df["int_rate"].astype(float)
```

Cleaning up the emp_length column to have a numeric representation of the employment years of loan applicant. However we need to impute null values with mode here. As the percentage of null values is less and we are assigning the most frequent value, imputation is safe

```python
# Imputation of null values with mode value - Most frequent value to be on top, hence picking the 0th index
df["emp_length"] = df["emp_length"].fillna(df["emp_length"].mode()[0])
# Cleaning the values
df["emp_length"] = df["emp_length"].apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0]))
df["emp_length"] = df["emp_length"].astype(int)
```

Final column dropping to remove everything on customer behaviour - as this information will not be available when a new applicant approach for loan

```python
# Curated list of relevant columns based on data dictionary
relevant_features = ['addr_state', 
 'annual_inc', 
 'dti', 
 'emp_length', 
 'emp_title', 
 'funded_amnt', 
 'funded_amnt_inv', 
 'grade', 
 'home_ownership', 
 'installment', 
 'int_rate', 
 'issue_d', 
 'loan_amnt', 
 'loan_status', 
 'purpose', 
 'sub_grade', 
 'term', 
 'verification_status']

df = df.loc[:, relevant_features]
```

Add new feature from existing - extract year from date column

```python
import datetime
df['loan_start_year'] = df['issue_d'].apply(lambda x : datetime.datetime.strptime(x, '%b-%y').year)
```

Plotting box plot to identify outliers on numerical columns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns only
num_cols = df.select_dtypes(include=['float', 'int']).columns

# Create box plots for each numerical column
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(col)
```

Its clear from the plot that annual_inc has very high number of outliers

Removing outliers from the dataset based on annual_inc column

```python
first_quantile = df["annual_inc"].quantile(0.25)
third_quantile = df["annual_inc"].quantile(0.75)
inter_quantile_range = third_quantile - first_quantile
# Standard formula to calculate the upper bound and lower bound
lower_bound = first_quantile - 1.5 * inter_quantile_range
upper_bound = third_quantile + 1.5 * inter_quantile_range
df = df[(df["annual_inc"] >= lower_bound) & (df["annual_inc"] <= upper_bound)]
```

Shape of final dataframe to be used for analysis

```python
df.shape
```

```python
for col in num_cols:
    sns.histplot(x=col, data=df, kde=False)
    plt.xticks(rotation=90)
    plt.title(col)
    plt.show()
```

Some intersting insights based on the distribution of above plot:
1. Large number of loan applicants has less than 10 years of experience
2. Annual income of the applicants is bit right skewed as majority of them earn below 80k.
3. DTI provides a perfect bell curve representing normal distribution


Visualizing Loan Status

```python
axis = sns.countplot(x="loan_status", data=df)
ncount = len(df)
axis.set_xticklabels(axis.get_xticklabels(), ha="center", fontsize=12)
axis_2 = axis.twinx()
plt.tight_layout()
for patch in axis.patches:
    x=patch.get_bbox().get_points()[:,0]
    y=patch.get_bbox().get_points()[1,1]
    axis.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), ha='center', va='bottom') 
axis_2.grid(None)
```

14.8% of the loan applicant has defaulted

Visualising years

```python
axis = sns.countplot(x="loan_start_year", data=df)
ncount = len(df)
axis.set_xticklabels(axis.get_xticklabels(), ha="center", fontsize=12)
axis_2 = axis.twinx()
plt.tight_layout()
for patch in axis.patches:
    x=patch.get_bbox().get_points()[:,0]
    y=patch.get_bbox().get_points()[1,1]
    axis.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), ha='center', va='bottom') 
axis_2.grid(None)
```

Every year the number pf application has increased

Effect of experience on loan status

```python
fig, ax = plt.subplots(figsize = (6,4))
sns.countplot(x='emp_length', data=df[df['loan_status'] == 'Charged Off'])
plt.xticks(rotation=90)
plt.show()
```

Its clear from above plot that applicants with less experience do tend to default more, because if the count of less experiences are stacked together, it will be clearly taller compared to 10+ years

Checking which credit grade applicant defaulted the most

```python
# Feature engineer grade column to represent them into numerical values first
grade_map = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7}
df['grade_num'] = df['grade'].map(grade_map)

fig, ax = plt.subplots(figsize = (6,4))
sns.countplot(x='grade_num', data=df[df['loan_status'] == 'Charged Off'])
plt.xticks(rotation=90)
plt.show()
```

From above plot, its clear that grade B. C and D applicants tend to default the most

Now lets analyse the grade and interest rate column, as grade could define how high or low the interest rate would be

```python
#Get correlation score
corr_score = df['grade_num'].corr(df['int_rate'])

#Plotting the correlation using scatterplot
sns.scatterplot(data=df, x='grade', y='int_rate')
plt.text(0.5, 0.9, 'Correlation Score: {}'.format(corr_score), transform=plt.gca().transAxes)
plt.xticks(rotation=90)
plt.show()
```

As expected, grade and interest rate are positively correlated

Plotting a heatmap for correlation to get a complete picture

```python
# Generate correlation score
corr = df.corr()
# Define size of the plot
plt.figure(figsize=(16, 10))
sns.set(font_scale=1.2)
# Plotting heatmap
sns.heatmap(corr, cmap='coolwarm', annot=True, annot_kws={'size': 12}, fmt='.2f', square=True)
plt.title('Correlation Matrix')
plt.show()
```
